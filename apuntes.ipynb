# Flexbox y CSS Grid OpenWebinars

En este curso aprenderé:

* Maquetación usando contenedores CSS FLEX.
* Maquetación usando contenedores CSS GRID.
* Comprenderé cuando elegir uno u otro.

En este curso no aprenderé:

* Maquetación web con las técnicas tradicionales (elementos flotantes, etc...)
* Diseño responsivo.
* Efectos y animaciones con CSS3.

FLEX y GRID son dos nuevos valores (HTML5) que vamos a poder dar a la propiedad CSS display y que nos van a permitir, junto con otras propiedades, maquetar nuestras páginas web de una manera más fácil a la que se usaba tradicionalmente.

## FLEX

* Elemento contenedor con display:flex
* Podremos: alinear, dar tamaño, distribuir el espacio restante.
* En una sola dirección

A todos los elementos que estén dentro sin saber ni siquiera que tamaño tienen.

## GRID

* Elemento contenedor con display:grid
* Es el sistema para maquetación más potente 
* Me permite trabajar con filas y columnas (no una sóla dimensión)

## TRABAJANDO CON FLEX

* La idea principal de la maquetación con elementos FLEX es que vamos a tener un elemento (una etiqueta) que va a poder controlar propiedades de los elementos que contiene.

### Elementos de Maquetación FLEX

* El contenedor FLEX (display:flex) --> Etiqueta padre
* Los elementos flexibles que están dentro y cuyas propiedades modificaremos --> Etiquetas hijas

### Propiedades Modificables

* La altura
* La anchura
* El orden
* La alineación vertical
* La alineación horizontal
* La distribución a lo largo del contenedor 

Es decir, propiedades que usamos para maquetar y de una manera más fácil a la forma tradicional.

### Dirección de los Elementos Flexibles (flex-direction)

* row (default)
* row-reverse
* column
* column-reverse

### Ajuste de los Elementos Flexibles si no caben (flex-wrap)

* nowrap (default) --> Una sola línea
* wrap --> lo elementos flexibles se disponen en múltiples líneas si no caben. De arriba abajo.
* wrap-reverse --> lo mismo que el anterior pero de abajo a arriba.

### Juntando la Dirección y el Ajuste (flex-flow)

* flex-flow:flex-direction flex-wrap

### Alineación Horizontal (justify-content)

* flex-start
* flex-end
* center
* space-between
* space-around
* space-evenly

### Alineación Vertical (align-items)

* flex-start
* flex-end
* center
* stretch
* baseline

### Alineación Vertical varias líneas - Wrap (align-content)

* flex-start
* flex-end
* center
* stretch
* space-between
* space-around

## ELEMENTOS FLEXIBLES: TAMAÑO Y ORDEN

### Orden (order)

* Podemos modificar el orden en el que se van a mostrar los elementos flexibles añadiéndoles la propiedad CSS order (valor por defecto 0)

* Cuanto mayor es el valor antes se muestra el elemento, en caso de empate el que estuviera antes en el código.

### Tamaño

* flex-grow: el factor de crecimiento de un elemento flexible cuando se distribuye el espacio restante (por defecto 1 pero debo indicarlo en todos).

* flex-shrink: el factor de contracción de un elemento flexible cuando sobrepasan el tamaño del contenedor (por defecto 1 pero debo indicarlo en todos).

* flex-basis: tamaño de un elemento antes de que el espacio restante (negativo o positivo) se distribuye (auto por defecto).

* todo junto --> flex:grow shrink flex-basis-value

### Alineación Vertical Individual (align-self)

* Desde el contenedor FLEX podíamos controlar la alineación vertical de todos los elementos flexibles a la vez con la propiedad align-items.

* Si queremos que un elemento tenga una alineación propia debemos usar la propiedad align-self.

## TRABAJANDO CON GRID

* La idea principal de la maquetación con elemento GRID es que vamos a tener un elemento (una etiqueta) que va a poder controlar propiedades de los elementos y establecer estructuras complejas para distribuirlos.

### Elementos de Maquetación GRID

* El contenedor GRID (display:grid o display:inline-grid)

* Los elementos grid que están dentro, cuyas propiedades modificaremos y que colocaremos atendiendo a una estructura compleja.

### Propiedades Contenedor GRID 

* Estructura en filas y columnas y la separación entre ellas.

* Definir áreas del GRID con nombre.

* Las alineaciones horizontal y vertical de los elementos del GRID y del propio GRID.

Es decir propiedades que usamos para maquetar y de una forma más fácil a la forma tradicional.

### Definir Estructura del GRID

* grid-template-columns: Número de columnas del grid y tamaño de las mismas.

* grid-template-rows: Número de filas del grid y tamaño de las mismas.

* grid-row-gap: Separación entre filas.

* grid-column-gap: Separación entre las columnas.

### Columnas

* grid-template-columns: Debemos de poner tantos valores de anchura como columnas quiero que tenga el GRID.

grid-template-columns: 20% 50% 30%
grid-template-columns: 100px auto 100px 100px
grid-template-columns: auto auto auto auto auto
grid-template-columns: [id] 100px auto [nombre] 300px [apellidos] auto

### Filas

* grid-template-rows: Debemos de poner tantos valores de altura como filas quiero que tenga el GRID.

grid-template-rows: 20% 50% 30%
grid-template-rows: 100px auto 100px 100px
grid-template-rows: auto auto auto auto auto
grid-template-rows: [uno] 100px [dos] 300px [tres] auto

### Repetir Valores y Unidad FR

* grid-template-column: repeat(3,20% [col-start]) auto

* grid-template-column: 2fr 100px 1fr 2fr

### Alineación Horizontal (justify-items)

* Para alinear horizontalmente el contenido que hay dentro de cada celda usaré la propiedad CSS justify-items.

* start
* end
* center
* stretch (por defecto)

### Alineación Vertical (align-items)

* start
* end
* center
* strecth (por defecto)

### Juntar Alineación Horizontal y Vertical

* place-item: start(vertical) end(horizontal)

### Distribución Dentro del Contenedor

* En el caso de que el GRID no ocupe todo el contenedor GRID podemos distribuirlo con las propiedades CSS justify-content (horizontal) y align-content (vertical)

* start
* end
* center
* stretch
* space-between
* space-around
* justify-content
* space-evenly

## ELEMENTOS DEL GRID

### Propiedades Elementos Grid

* Área del GRID que va a ocupar.

* Alineación horizontal de manera individual.

* Alineación vertical de manera individual.

* Colocación implícita si no especificamos nada.

No les afecta: float, display:inline-block, display-table-cell, vertical-align o column-*

### Especificar el Area a Ocupar

* grid-column-start
* grid-column-end
* grid-row-start
* grid-row-end

Podemos juntar estas propiedades

* grid-column: start / end
* grid-row: start / end
* grid: row-start column-start row-end column-end

### Alineación Horizontal Individual (justify-self)

* start
* end
* center
* stretch (por defecto)

Los valores son los mismos que usábamos en la propiedad justify-items del contenedor grid.

### Alineación Vertical Individual (align-self)

* start
* end
* center
* stretch (por defecto)

Los valores son los mismos que usábamos en la propiedad align-items del contenedor grid.

### Colocación Implícita 

* La colocación implícita de los elementos GRID es lo que sucede cuando colocamos elementos GRID fuera de la estructura que se ha definido en el contenedor GRID o cuando no les damos pocisión.

* grid-auto-columns
* grid-auto-rows
* grid-auto-flow

Son propiedades que damos al CONTENEDOR.

* Si tengo elementos sin definir una colocación en el contenedor GRID la propiedad del contenedor grid-auto-flow define cómo se van a colocar esos elementos.

### Valores para grid-auto-flow

* row: rellena las filas primero (opción por defecto)

* column: rellena las columnas primero

* dense: intenta rellenar primero los huecos si vienen elementos más pequeños. Puede cambiar el orden de los elementos así que hay que tener cuidado

### Áreas GRID

* Si no queremos especificar el tamaño para cada elemento del GRID podemos darles nombre a su propiedad gird-area y usar esos nombres en la propiedad grid-template-area

## FLEX O GRID

* Independientemente de qué sistema utilicemos, maquetar con FLEX o con GRID es mucho más fácil que maquetar usando sistemas tradicionales basados únicamente en float, position, tablas, etc...

* Características de FLEX:
    * Ventajas:
        * Una sola dirección
        * Controlo fácilmente la alineación de los elementos flexibles
    * Desventajas:
        * Para el mismo layout que GRID necesito una estructura HTML más compleja

* Características de GRID:
    * Ventajas:
        * Dos dimensiones
        * Puedo definir muy fácilmente la estructura general
        * Es muy fácil definir la separación (gap) entre las celdas.
    * Desventajas:
        * La alineación "dentro" de las celdas

* Conclusión: hay cosas que sólo puedo hacer con Flex (ej: alineación de los elementos dentro de una zona) y hay cosas que sólo puedo hacer con Grid (ej: establecer una estructura a nivel general para toda la página).

USA LAS DOS, nada impide que un área de un contenedor GRID sea a suvez un contenedor FLEX.
